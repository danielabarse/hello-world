```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import solve_ivp, odeint
```

1. Resolver numéricamente con `solve_ivp` la siguiente ecuación diferencial:
$$\frac{dx}{dt}=\frac{x}{t+1}$$

```python
def derivada(t, x, r):
    return x/(t+1)
tspan = [0, 20]
x0 = np.array([0.5])
r = 0.5

x = solve_ivp(derivada,tspan, x0, args=(r,), max_step = 0.1)
x
```

```python
plt.figure()
plt.plot(x.t, x.y[0,:], label="Solución numérica")
plt.legend(loc="best")
plt.grid()
```

2. Encuentre la solución numérica para un oscilador armónico simple:

$$
\frac{d^2 x}{dt^2} +  \omega_{0}^2 x = 0.
$$
Donde $x_0=0.5$, $\dot{x}_0=0.1$ y $\omega_0=3$.

2.1 Defina una función, cuyos argumentos de entrada sean:


1. el tiempo $t$,
2. el arreglo $[x,\dot{x}]$.

Y la salida:

1. el arreglo $[\dot{x},\ddot{x}]$.

  Considerando el argumento de la frecuencia natural $\omega_0$ debe ser una variable global.

```python
def g(t, X):
    x = X[0]
    dx = X[1]
    dx2 = - (w0**2)*x
    return [dx, dx2]
```

```python
tspan = [0, 10]
w0 = 3 
x0 = np.array([0.5, 0.1])

x = solve_ivp(g ,tspan, x0, max_step = 0.1)

plt.figure()
plt.plot(x.t, x.y[0,:], label="Posición")
plt.plot(x.t, x.y[1,:], label="Velocidad")
plt.legend(loc="best")
plt.grid()
```
